```python
import seaborn as sns
import numpy as py
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset("Titanic")
sns.boxplot(x = "survived",
           y = "age",showmeans = True,
                 meanprops = {"marker" : "*",
                             "markersize":"12",
                             "markeredgecolor":"red"},
            data=kashti)
#show labe

plt.xlabel('How many survived',size = 12)
plt.ylabel('Age (age)',size = 12) 
plt.title("Box  plot for kitne doobe aur kitne batche",size = 14,weight = 'bold')
plt.show()
```

```python
tip = sns.load_dataset("tips")
tip.head()
```

```python
g = sns.FacetGrid(tip,col="time",row="sex")
g.map(sns.scatterplot,"total_bill","tip")
plt.show()
```

```python
g = sns.FacetGrid(tip,col="time",row="sex")
g.map_dataframe(sns.histplot,"total_bill")
plt.show()
```
